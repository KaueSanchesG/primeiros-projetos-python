## Problema a ser resolvido - passo a passo
- Buscar automaticamente os dados das ações
- Gerar as análises de forma automática
- Enviar um e-mail para o gestor

# Passo 1 - Buscar automaticamente os dados das ações


```python
!pip install yfinance
```

```python
import yfinance
```

```python
codigo = input("Digite o codigo da ação desejada: ")
dados = yfinance.Ticker(str(codigo)).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

## Passo 2 - Gerar as análises de forma automática

- Análise dos últimos seis meses
- Cotação miníma
- Cotação atual

```python
cot_max = fechamento.max()
cot_min = fechamento.min()
cot_atual = fechamento[-1]
```

## Passo 3 - Enviar um e-mail para o nosso gestor

- Abrir uma nova aba no navegador
- Digitar o endereço do gmail e dar um enter
- Clicar no botão escrever
- Colocar o destinatário (dar um tab)
- Colocar o assunto (dar um tab)
- COlocar o conteúdo
- Clicar no botão enviar

```python
!pip install pyautogui
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 3

# abrir uma nova aba no navegador
pyautogui.hotkey("ctrl", "t")

# Digitar o endereço do gmail e dar um enter
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# Clicar no botão escrever
pyautogui.click(x=97, y=241)

# Colocar o destinatário (dar um tab)
pyperclip.copy("goncaleskamillyvictoria@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Colocar o assunto (dar um tab)
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

mensagem = f"""
Prezado,

Seguem as análises diárias dos últimos 6(seis) meses da ação {codigo}:

Cotação máxima: R${round(cot_max, 2)}
Cotação mínima: R${round(cot_min, 2)}
Cotação atual: R${round(cot_atual, 2)}

Em caso de dúvidas fico à disposição,

Grato!
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Click no enviar

pyautogui.hotkey("ctrl", "enter")
```
